```python
import pandas as pd
import numpy as np
import string


tickers = list(string.ascii_uppercase)
sp1 = pd.DataFrame(index=tickers, columns = ["rnk"])
sp1.index.name = "ticker"
rnks = np.array(range(1, 27))
np.random.shuffle(rnks)
sp1.rnk = rnks
sp1.to_excel("Ranks.xlsx")
```

```python
sp2 = sp1.copy()
sp2["price"] = np.random.uniform(low=10, high=100, size=26)
sp2.price = sp2.price.round(2)
sp2 = sp2.drop(columns=["rnk"])
sp2.to_excel("Prices.xlsx")
```
